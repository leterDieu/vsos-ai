```python
from datetime import date, datetime
import json

from pandas.testing import assert_frame_equal

EXPECTED_FAVORITE = (1, ["S0167"])
EXPECTED_COVERAGE = ["M011"]
EXPECTED_PEAK = {"peak_count": 11, "peak_moment": "2025-03-06 10:53"}
EXPECTED_MAX_MOVIES = 4
EXPECTED_STREAK = 15
EXPECTED_RANGE = 21
EXPECTED_PLAYLIST = {"total_enjoyment": 281, "playlist": ["M001", "M004", "M010"]}

EXPECTED_INTERVAL_COUNTS_RAW = json.loads(
    '''
    [
      {
        "interval_start": "2025-03-13T00:00:00",
        "interval_end": "2025-03-13T00:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T06:00:00",
        "interval_end": "2025-03-13T06:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T06:30:00",
        "interval_end": "2025-03-13T07:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T07:00:00",
        "interval_end": "2025-03-13T07:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T07:30:00",
        "interval_end": "2025-03-13T08:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T08:00:00",
        "interval_end": "2025-03-13T08:30:00",
        "viewer_count": 2
      },
      {
        "interval_start": "2025-03-13T08:30:00",
        "interval_end": "2025-03-13T09:00:00",
        "viewer_count": 3
      },
      {
        "interval_start": "2025-03-13T09:00:00",
        "interval_end": "2025-03-13T09:30:00",
        "viewer_count": 3
      },
      {
        "interval_start": "2025-03-13T09:30:00",
        "interval_end": "2025-03-13T10:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T10:00:00",
        "interval_end": "2025-03-13T10:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T10:30:00",
        "interval_end": "2025-03-13T11:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T11:00:00",
        "interval_end": "2025-03-13T11:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T11:30:00",
        "interval_end": "2025-03-13T12:00:00",
        "viewer_count": 2
      },
      {
        "interval_start": "2025-03-13T12:00:00",
        "interval_end": "2025-03-13T12:30:00",
        "viewer_count": 3
      },
      {
        "interval_start": "2025-03-13T12:30:00",
        "interval_end": "2025-03-13T13:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T13:00:00",
        "interval_end": "2025-03-13T13:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T13:30:00",
        "interval_end": "2025-03-13T14:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T14:00:00",
        "interval_end": "2025-03-13T14:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T14:30:00",
        "interval_end": "2025-03-13T15:00:00",
        "viewer_count": 3
      },
      {
        "interval_start": "2025-03-13T15:00:00",
        "interval_end": "2025-03-13T15:30:00",
        "viewer_count": 2
      },
      {
        "interval_start": "2025-03-13T15:30:00",
        "interval_end": "2025-03-13T16:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T16:00:00",
        "interval_end": "2025-03-13T16:30:00",
        "viewer_count": 2
      },
      {
        "interval_start": "2025-03-13T16:30:00",
        "interval_end": "2025-03-13T17:00:00",
        "viewer_count": 6
      },
      {
        "interval_start": "2025-03-13T17:00:00",
        "interval_end": "2025-03-13T17:30:00",
        "viewer_count": 7
      },
      {
        "interval_start": "2025-03-13T17:30:00",
        "interval_end": "2025-03-13T18:00:00",
        "viewer_count": 9
      },
      {
        "interval_start": "2025-03-13T18:00:00",
        "interval_end": "2025-03-13T18:30:00",
        "viewer_count": 7
      },
      {
        "interval_start": "2025-03-13T18:30:00",
        "interval_end": "2025-03-13T19:00:00",
        "viewer_count": 3
      },
      {
        "interval_start": "2025-03-13T19:00:00",
        "interval_end": "2025-03-13T19:30:00",
        "viewer_count": 2
      },
      {
        "interval_start": "2025-03-13T19:30:00",
        "interval_end": "2025-03-13T20:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T20:00:00",
        "interval_end": "2025-03-13T20:30:00",
        "viewer_count": 2
      },
      {
        "interval_start": "2025-03-13T20:30:00",
        "interval_end": "2025-03-13T21:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T21:00:00",
        "interval_end": "2025-03-13T21:30:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T21:30:00",
        "interval_end": "2025-03-13T22:00:00",
        "viewer_count": 1
      },
      {
        "interval_start": "2025-03-13T22:00:00",
        "interval_end": "2025-03-13T22:30:00",
        "viewer_count": 3
      },
      {
        "interval_start": "2025-03-13T22:30:00",
        "interval_end": "2025-03-13T23:00:00",
        "viewer_count": 5
      },
      {
        "interval_start": "2025-03-13T23:00:00",
        "interval_end": "2025-03-13T23:30:00",
        "viewer_count": 5
      },
      {
        "interval_start": "2025-03-13T23:30:00",
        "interval_end": "2025-03-14T00:00:00",
        "viewer_count": 4
      }
    ]
    '''
)

EXPECTED_COMPLETION_RAW = json.loads(
    '''
    [
      {
        "user_id": "U001",
        "movie_id": "M001",
        "completion_ratio": 1.0,
        "unique_minutes": 118.0
      },
      {
        "user_id": "U001",
        "movie_id": "M002",
        "completion_ratio": 1.0,
        "unique_minutes": 105.0
      },
      {
        "user_id": "U001",
        "movie_id": "M003",
        "completion_ratio": 0.65625,
        "unique_minutes": 63.0
      },
      {
        "user_id": "U001",
        "movie_id": "M004",
        "completion_ratio": 1.0,
        "unique_minutes": 83.0
      },
      {
        "user_id": "U001",
        "movie_id": "M005",
        "completion_ratio": 1.0,
        "unique_minutes": 132.0
      },
      {
        "user_id": "U001",
        "movie_id": "M006",
        "completion_ratio": 1.0,
        "unique_minutes": 114.0
      },
      {
        "user_id": "U001",
        "movie_id": "M007",
        "completion_ratio": 1.0,
        "unique_minutes": 101.0
      },
      {
        "user_id": "U001",
        "movie_id": "M008",
        "completion_ratio": 1.0,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U001",
        "movie_id": "M009",
        "completion_ratio": 1.0,
        "unique_minutes": 109.0
      },
      {
        "user_id": "U001",
        "movie_id": "M010",
        "completion_ratio": 1.0,
        "unique_minutes": 95.0
      },
      {
        "user_id": "U001",
        "movie_id": "M011",
        "completion_ratio": 1.0,
        "unique_minutes": 88.0
      },
      {
        "user_id": "U001",
        "movie_id": "M012",
        "completion_ratio": 1.0,
        "unique_minutes": 122.0
      },
      {
        "user_id": "U002",
        "movie_id": "M001",
        "completion_ratio": 1.0,
        "unique_minutes": 118.0
      },
      {
        "user_id": "U002",
        "movie_id": "M002",
        "completion_ratio": 1.0,
        "unique_minutes": 105.0
      },
      {
        "user_id": "U002",
        "movie_id": "M003",
        "completion_ratio": 1.0,
        "unique_minutes": 96.0
      },
      {
        "user_id": "U002",
        "movie_id": "M004",
        "completion_ratio": 1.0,
        "unique_minutes": 83.0
      },
      {
        "user_id": "U002",
        "movie_id": "M005",
        "completion_ratio": 0.9393939393939394,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U002",
        "movie_id": "M006",
        "completion_ratio": 1.0,
        "unique_minutes": 114.0
      },
      {
        "user_id": "U002",
        "movie_id": "M007",
        "completion_ratio": 1.0,
        "unique_minutes": 101.0
      },
      {
        "user_id": "U002",
        "movie_id": "M008",
        "completion_ratio": 1.0,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U002",
        "movie_id": "M009",
        "completion_ratio": 1.0,
        "unique_minutes": 109.0
      },
      {
        "user_id": "U002",
        "movie_id": "M010",
        "completion_ratio": 1.0,
        "unique_minutes": 95.0
      },
      {
        "user_id": "U002",
        "movie_id": "M011",
        "completion_ratio": 1.0,
        "unique_minutes": 88.0
      },
      {
        "user_id": "U002",
        "movie_id": "M012",
        "completion_ratio": 1.0,
        "unique_minutes": 122.0
      },
      {
        "user_id": "U003",
        "movie_id": "M001",
        "completion_ratio": 1.0,
        "unique_minutes": 118.0
      },
      {
        "user_id": "U003",
        "movie_id": "M002",
        "completion_ratio": 1.0,
        "unique_minutes": 105.0
      },
      {
        "user_id": "U003",
        "movie_id": "M003",
        "completion_ratio": 1.0,
        "unique_minutes": 96.0
      },
      {
        "user_id": "U003",
        "movie_id": "M004",
        "completion_ratio": 1.0,
        "unique_minutes": 83.0
      },
      {
        "user_id": "U003",
        "movie_id": "M005",
        "completion_ratio": 0.8484848484848485,
        "unique_minutes": 112.0
      },
      {
        "user_id": "U003",
        "movie_id": "M006",
        "completion_ratio": 1.0,
        "unique_minutes": 114.0
      },
      {
        "user_id": "U003",
        "movie_id": "M007",
        "completion_ratio": 1.0,
        "unique_minutes": 101.0
      },
      {
        "user_id": "U003",
        "movie_id": "M008",
        "completion_ratio": 1.0,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U003",
        "movie_id": "M009",
        "completion_ratio": 1.0,
        "unique_minutes": 109.0
      },
      {
        "user_id": "U003",
        "movie_id": "M010",
        "completion_ratio": 1.0,
        "unique_minutes": 95.0
      },
      {
        "user_id": "U003",
        "movie_id": "M011",
        "completion_ratio": 1.0,
        "unique_minutes": 88.0
      },
      {
        "user_id": "U003",
        "movie_id": "M012",
        "completion_ratio": 1.0,
        "unique_minutes": 122.0
      },
      {
        "user_id": "U004",
        "movie_id": "M001",
        "completion_ratio": 1.0,
        "unique_minutes": 118.0
      },
      {
        "user_id": "U004",
        "movie_id": "M002",
        "completion_ratio": 1.0,
        "unique_minutes": 105.0
      },
      {
        "user_id": "U004",
        "movie_id": "M003",
        "completion_ratio": 1.0,
        "unique_minutes": 96.0
      },
      {
        "user_id": "U004",
        "movie_id": "M004",
        "completion_ratio": 1.0,
        "unique_minutes": 83.0
      },
      {
        "user_id": "U004",
        "movie_id": "M005",
        "completion_ratio": 0.9166666666666666,
        "unique_minutes": 121.0
      },
      {
        "user_id": "U004",
        "movie_id": "M006",
        "completion_ratio": 1.0,
        "unique_minutes": 114.0
      },
      {
        "user_id": "U004",
        "movie_id": "M007",
        "completion_ratio": 1.0,
        "unique_minutes": 101.0
      },
      {
        "user_id": "U004",
        "movie_id": "M008",
        "completion_ratio": 1.0,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U004",
        "movie_id": "M009",
        "completion_ratio": 1.0,
        "unique_minutes": 109.0
      },
      {
        "user_id": "U004",
        "movie_id": "M010",
        "completion_ratio": 1.0,
        "unique_minutes": 95.0
      },
      {
        "user_id": "U004",
        "movie_id": "M011",
        "completion_ratio": 1.0,
        "unique_minutes": 88.0
      },
      {
        "user_id": "U004",
        "movie_id": "M012",
        "completion_ratio": 1.0,
        "unique_minutes": 122.0
      },
      {
        "user_id": "U005",
        "movie_id": "M001",
        "completion_ratio": 1.0,
        "unique_minutes": 118.0
      },
      {
        "user_id": "U005",
        "movie_id": "M002",
        "completion_ratio": 1.0,
        "unique_minutes": 105.0
      },
      {
        "user_id": "U005",
        "movie_id": "M003",
        "completion_ratio": 1.0,
        "unique_minutes": 96.0
      },
      {
        "user_id": "U005",
        "movie_id": "M004",
        "completion_ratio": 1.0,
        "unique_minutes": 83.0
      },
      {
        "user_id": "U005",
        "movie_id": "M005",
        "completion_ratio": 1.0,
        "unique_minutes": 132.0
      },
      {
        "user_id": "U005",
        "movie_id": "M006",
        "completion_ratio": 1.0,
        "unique_minutes": 114.0
      },
      {
        "user_id": "U005",
        "movie_id": "M007",
        "completion_ratio": 1.0,
        "unique_minutes": 101.0
      },
      {
        "user_id": "U005",
        "movie_id": "M008",
        "completion_ratio": 1.0,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U005",
        "movie_id": "M009",
        "completion_ratio": 1.0,
        "unique_minutes": 109.0
      },
      {
        "user_id": "U005",
        "movie_id": "M010",
        "completion_ratio": 1.0,
        "unique_minutes": 95.0
      },
      {
        "user_id": "U005",
        "movie_id": "M011",
        "completion_ratio": 1.0,
        "unique_minutes": 88.0
      },
      {
        "user_id": "U005",
        "movie_id": "M012",
        "completion_ratio": 1.0,
        "unique_minutes": 122.0
      },
      {
        "user_id": "U006",
        "movie_id": "M001",
        "completion_ratio": 1.0,
        "unique_minutes": 118.0
      },
      {
        "user_id": "U006",
        "movie_id": "M002",
        "completion_ratio": 1.0,
        "unique_minutes": 105.0
      },
      {
        "user_id": "U006",
        "movie_id": "M003",
        "completion_ratio": 1.0,
        "unique_minutes": 96.0
      },
      {
        "user_id": "U006",
        "movie_id": "M004",
        "completion_ratio": 1.0,
        "unique_minutes": 83.0
      },
      {
        "user_id": "U006",
        "movie_id": "M005",
        "completion_ratio": 1.0,
        "unique_minutes": 132.0
      },
      {
        "user_id": "U006",
        "movie_id": "M006",
        "completion_ratio": 1.0,
        "unique_minutes": 114.0
      },
      {
        "user_id": "U006",
        "movie_id": "M007",
        "completion_ratio": 1.0,
        "unique_minutes": 101.0
      },
      {
        "user_id": "U006",
        "movie_id": "M008",
        "completion_ratio": 1.0,
        "unique_minutes": 124.0
      },
      {
        "user_id": "U006",
        "movie_id": "M009",
        "completion_ratio": 1.0,
        "unique_minutes": 109.0
      },
      {
        "user_id": "U006",
        "movie_id": "M010",
        "completion_ratio": 1.0,
        "unique_minutes": 95.0
      },
      {
        "user_id": "U006",
        "movie_id": "M011",
        "completion_ratio": 1.0,
        "unique_minutes": 88.0
      },
      {
        "user_id": "U006",
        "movie_id": "M012",
        "completion_ratio": 1.0,
        "unique_minutes": 122.0
      }
    ]
    '''
)

EXPECTED_CHRONOTYPE_RAW = json.loads(
    '''
    [
      {
        "user_id": "U001",
        "night_share": 0.017943,
        "morning_share": 0.289971,
        "day_share": 0.411086,
        "evening_share": 0.281,
        "centroid_time": "14:52"
      },
      {
        "user_id": "U002",
        "night_share": 0.027397,
        "morning_share": 0.34551,
        "day_share": 0.277847,
        "evening_share": 0.349246,
        "centroid_time": "15:10"
      },
      {
        "user_id": "U003",
        "night_share": 0.003262,
        "morning_share": 0.28418,
        "day_share": 0.347649,
        "evening_share": 0.364909,
        "centroid_time": "15:25"
      },
      {
        "user_id": "U004",
        "night_share": 0.002857,
        "morning_share": 0.347571,
        "day_share": 0.299571,
        "evening_share": 0.35,
        "centroid_time": "15:06"
      },
      {
        "user_id": "U005",
        "night_share": 0.008127,
        "morning_share": 0.234883,
        "day_share": 0.514304,
        "evening_share": 0.242685,
        "centroid_time": "15:06"
      },
      {
        "user_id": "U006",
        "night_share": 0.010268,
        "morning_share": 0.386452,
        "day_share": 0.435791,
        "evening_share": 0.167489,
        "centroid_time": "13:59"
      }
    ]
    '''
)

EXPECTED_RETENTION_RAW = json.loads(
    '''
    [
      {
        "user_id": "U001",
        "max_minutes_7d": 3466,
        "max_active_days_7d": 7
      },
      {
        "user_id": "U002",
        "max_minutes_7d": 3963,
        "max_active_days_7d": 7
      },
      {
        "user_id": "U003",
        "max_minutes_7d": 4054,
        "max_active_days_7d": 7
      },
      {
        "user_id": "U004",
        "max_minutes_7d": 3336,
        "max_active_days_7d": 7
      },
      {
        "user_id": "U005",
        "max_minutes_7d": 3053,
        "max_active_days_7d": 7
      },
      {
        "user_id": "U006",
        "max_minutes_7d": 3841,
        "max_active_days_7d": 7
      }
    ]
    '''
)

EXPECTED_INTERVAL_COUNTS = pd.DataFrame(EXPECTED_INTERVAL_COUNTS_RAW)
EXPECTED_INTERVAL_COUNTS["interval_start"] = pd.to_datetime(EXPECTED_INTERVAL_COUNTS["interval_start"])
EXPECTED_INTERVAL_COUNTS["interval_end"] = pd.to_datetime(EXPECTED_INTERVAL_COUNTS["interval_end"])
EXPECTED_INTERVAL_COUNTS = EXPECTED_INTERVAL_COUNTS.sort_values(["interval_start", "interval_end"]).reset_index(drop=True)

EXPECTED_COMPLETION = pd.DataFrame(EXPECTED_COMPLETION_RAW).sort_values(["user_id", "movie_id"]).reset_index(drop=True)
EXPECTED_COMPLETION["completion_ratio"] = EXPECTED_COMPLETION["completion_ratio"].astype(float)
EXPECTED_COMPLETION["unique_minutes"] = EXPECTED_COMPLETION["unique_minutes"].astype(float)

EXPECTED_CHRONOTYPE = pd.DataFrame(EXPECTED_CHRONOTYPE_RAW).sort_values("user_id").reset_index(drop=True)
float_cols = ["night_share", "morning_share", "day_share", "evening_share"]
EXPECTED_CHRONOTYPE[float_cols] = EXPECTED_CHRONOTYPE[float_cols].astype(float)

EXPECTED_RETENTION = pd.DataFrame(EXPECTED_RETENTION_RAW).sort_values("user_id").reset_index(drop=True)

FULL_USER_DAILY_MINUTES = (
    sessions_df.assign(date=sessions_df["start_time"].dt.date)
    .groupby(["user_id", "date"], as_index=False)["watched_minutes"]
    .sum()
    .sort_values(["user_id", "date"])
)

FULL_USER_DAYS = {
    user: list(dates)
    for user, dates in FULL_USER_DAILY_MINUTES.groupby("user_id")["date"].apply(list).items()
}

U002_DURATIONS = (
    sessions_df.query("user_id == 'U002'").sort_values("start_time")["watched_minutes"].tolist()
)


def run_test(name: str, fn):
    try:
        fn()
        print(f"[OK] {name}")
    except NotImplementedError:
        print(f"[SKIP] {name} - функция не реализована")
    except AssertionError as err:
        print(f"[FAIL] {name}: {err}")


def build_plan_and_extras(user_id: str, day: str) -> tuple[pd.DataFrame, pd.DataFrame, int, int]:
    subset = sessions_df[
        (sessions_df["user_id"] == user_id)
        & (sessions_df["start_time"].dt.date == pd.to_datetime(day).date())
    ].copy()
    assert not subset.empty, "В датасете нет ожидаемых сессий для проверки"

    def to_minutes(ts: pd.Timestamp) -> int:
        return ts.hour * 60 + ts.minute

    plan_df = subset[subset["favorite_film"]][["movie_id", "start_time", "end_time"]].copy()
    extras_df = subset[~subset["favorite_film"]][["movie_id", "start_time", "end_time"]].copy()

    plan_df["start_min"] = plan_df["start_time"].map(to_minutes)
    plan_df["end_min"] = plan_df["end_time"].map(to_minutes)
    extras_df["start_min"] = extras_df["start_time"].map(to_minutes)
    extras_df["end_min"] = extras_df["end_time"].map(to_minutes)

    plan_df = plan_df[["movie_id", "start_min", "end_min"]]
    extras_df = extras_df[["movie_id", "start_min", "end_min"]]

    start_min = to_minutes(subset["start_time"].min())
    end_min = to_minutes(subset["end_time"].max())
    return plan_df, extras_df, start_min, end_min


def test_select_max_favorite_sessions_dataset():
    result = select_max_favorite_sessions(sessions_df, "U002", "2025-03-06")
    assert result == EXPECTED_FAVORITE


def test_cover_day_with_extras_dataset():
    plan_df, extras_df, start_min, end_min = build_plan_and_extras("U001", "2025-03-13")
    result = cover_day_with_extras(plan_df, extras_df, start_min, end_min)
    assert result == EXPECTED_COVERAGE


def test_find_peak_concurrency_dataset():
    assert find_peak_concurrency(sessions_df) == EXPECTED_PEAK


def test_build_interval_view_counts_dataset():
    actual = build_interval_view_counts(sessions_df, "2025-03-13", step_minutes=30)
    actual = actual.sort_values(["interval_start", "interval_end"]).reset_index(drop=True)
    assert_frame_equal(actual, EXPECTED_INTERVAL_COUNTS, check_dtype=False)


def test_max_movies_within_budget_dataset():
    assert max_movies_within_budget(U002_DURATIONS, limit=240) == EXPECTED_MAX_MOVIES


def test_longest_active_streak_dataset():
    assert longest_active_streak(FULL_USER_DAYS["U003"]) == EXPECTED_STREAK


def test_longest_window_within_range_dataset():
    assert longest_window_within_range(U002_DURATIONS, k=45) == EXPECTED_RANGE


def test_build_optimal_playlist_dataset():
    result = build_optimal_playlist(movies_df, time_limit=360)
    assert result == EXPECTED_PLAYLIST


def test_compute_completion_ratio_dataset():
    actual = compute_completion_ratio(sessions_df, movies_df)
    actual = actual.sort_values(["user_id", "movie_id"]).reset_index(drop=True)
    assert_frame_equal(actual, EXPECTED_COMPLETION, check_dtype=False, atol=1e-9, rtol=1e-9)


def test_compute_chronotype_features_dataset():
    actual = compute_chronotype_features(sessions_df)
    actual = actual.sort_values("user_id").reset_index(drop=True)
    rounded = actual.copy()
    rounded[float_cols] = rounded[float_cols].round(6)
    assert_frame_equal(rounded, EXPECTED_CHRONOTYPE, check_dtype=False)


def test_compute_retention_metrics_dataset():
    actual = compute_retention_metrics(FULL_USER_DAILY_MINUTES)
    actual = actual.sort_values("user_id").reset_index(drop=True)
    assert_frame_equal(actual, EXPECTED_RETENTION, check_dtype=False)


all_tests = [
    ("Greedy favorites full", test_select_max_favorite_sessions_dataset),
    ("Greedy coverage full", test_cover_day_with_extras_dataset),
    ("Peak concurrency full", test_find_peak_concurrency_dataset),
    ("Interval counts full", test_build_interval_view_counts_dataset),
    ("Two pointers 240 min full", test_max_movies_within_budget_dataset),
    ("Active streak full", test_longest_active_streak_dataset),
    ("Window with range full", test_longest_window_within_range_dataset),
    ("Knapsack playlist full", test_build_optimal_playlist_dataset),
    ("Completion ratio full", test_compute_completion_ratio_dataset),
    ("Chronotype full", test_compute_chronotype_features_dataset),
    ("Retention full", test_compute_retention_metrics_dataset),
]

for name, fn in all_tests:
    run_test(name, fn)
```
