## 1. Чтение CSV-файла

**Задача:** Прочитай файл `train.csv` и сохрани результат в переменную `df`.  
**Подсказка:** Используй функцию `pd.read_csv()`.

````python
import pandas as pd
````

**Если забыл `read_csv`:** Можно прочитать файл стандартным Python и создать DataFrame вручную:
```python
import csv

with open('train.csv') as f:
    reader = csv.DictReader(f)
    data = list(reader)

df = pd.DataFrame(data)
```
Это работает, но `read_csv` намного быстрее и сам определяет типы данных.

## 2. Первые N строк

**Задача:** Покажи первые 5 строк датафрейма.  
**Подсказка:** Метод `.head(n)` — по умолчанию `n=5`.

**Если забыл `.head()`:** Срез работает так же:
```python
df[:5]
# или
df.iloc[:5]
```

## 3. Статистика по числовым колонкам

**Задача:** Получи сводную статистику (count, mean, std, min, квартили, max) для всех числовых колонок.  
**Подсказка:** Метод `.describe()`.

**Если забыл `.describe()`:** Можно собрать статистику по отдельным функциям:
```python
print('Среднее:', df['temp'].mean())
print('Медиана:', df['temp'].median())
print('Мин:',     df['temp'].min())
print('Макс:',    df['temp'].max())
print('Стд:',     df['temp'].std())
print('Кол-во:',  df['temp'].count())
```
Это даёт тот же результат, но для одной колонки за раз.

---
## 4. Названия колонок

**Задача:** Выведи список всех колонок датафрейма.  
**Подсказка:** Атрибут `.columns`.

**Если забыл `.columns`:**
```python
list(df)          # возвращает список имён колонок
# или
df.keys()         # тоже работает
# или просто
df.head(0)        # покажет заголовки без данных
```

---
## 5. Типы данных колонок

**Задача:** Узнай типы данных каждой колонки.  
**Подсказка:** Атрибут `.dtypes`.

**Если забыл `.dtypes`:**
```python
df.info()       # выведет типы + кол-во непустых значений + использование памяти
# или для одной колонки:
type(df['temp'].iloc[0])   # тип первого элемента
```

---
## 6. Простая фильтрация (одно условие)

**Задача:** Получи все строки, где `workingday == 0` (выходные дни).  
**Подсказка:** Булева маска внутри `df[...]`.

**Если забыл синтаксис булевой маски:**
```python
df.query('workingday == 0')      # строковый синтаксис, легче запомнить
# или
df.loc[df['workingday'] == 0]    # то же самое, но через .loc
```

---
## 7. Логическое «И» — несколько условий одновременно

**Задача:** Получи строки, где `workingday == 1` **и** `season == 1` (весна в рабочие дни).  
**Подсказка:** Оператор `&`. Каждое условие **в скобках**.

**Если забыл `&`:**
```python
df.query('workingday == 1 and season == 1')   # через query — пишем как обычный Python
```

---
## 8. Логическое «ИЛИ»

**Задача:** Получи строки, где влажность < 10 **или** температура > 30.  
**Подсказка:** Оператор `|`. Каждое условие **в скобках**.

**Если забыл `|`:**
```python
df.query('humidity < 10 or temp > 30')
```

---
## 9. Логическое «НЕ»

**Задача:** Получи все строки, где `workingday` **не равен** 0.  
**Подсказка:** Оператор `~` перед условием в скобках.

**Если забыл `~`:** Проще всего — просто использовать `!=`:
```python
df[df['workingday'] != 0]
# или
df.query('workingday != 0')
```

---
## 10. Комбинация «НЕ» + «И»

**Задача:** Получи строки, где день **не** выходной, температура от 10 (включительно) до 30 (не включительно).  
**Подсказка:** `~`, `&` — каждое условие в скобках.

**Если забыл комбинацию операторов:**
```python
df.query('workingday != 0 and temp >= 10 and temp < 30')
# или с between:
df[(df['workingday'] != 0) & (df['temp'].between(10, 30, inclusive='left'))]
```

---
## 11. Проверка на пропуски — `.isna()`

**Задача:** Найди все строки, где `windspeed` пропущен (NaN).  
**Подсказка:** Метод `.isna()` на колонке.

**Если забыл `.isna()`:**
```python
df[df['windspeed'].isnull()]            # isnull() — синоним isna()
# или
df[pd.isna(df['windspeed'])]            # функция, а не метод
# или проверить целиком:
df[df.isnull().any(axis=1)]             # строки, где хоть один столбец пуст
```

---
## 12. Есть ли вообще пропуски?

**Задача:** Проверь, есть ли хотя бы один пропуск **во всём** датафрейме.  
**Подсказка:** `.isna().any().any()` — двойной `.any()` сначала по колонкам, потом в одно значение.

**Если забыл `.any().any()`:**
```python
df.isna().sum().sum()          # посчитает общее число пропусков числом
# или
df.isna().values.any()         # через numpy — одним вызовом
# или подробнее — сколько пропусков в каждой колонке:
df.isna().sum()                # Series с числом NaN по каждой колонке
```

---
## 13. Фильтрация по списку значений — `.isin()`

**Задача:** Получи строки, где `season` равен 1, 3 или 4.  
**Подсказка:** Метод `.isin([...])` на колонке.

**Если забыл `.isin()`:**
```python
df[(df['season'] == 1) | (df['season'] == 3) | (df['season'] == 4)]
# или
df.query('season in [1, 3, 4]')
# или исключить ненужное:
df[df['season'] != 2]
```

---
## 14. Доступ по индексу — `.loc` и `.iloc`

**Задача:** Получи строку с индексом 4.  
**Подсказка:** `.loc[метка]` — по значению индекса, `.iloc[номер]` — по порядковому номеру.

**Если забыл `.loc` / `.iloc`:**
```python
df.iloc[4]              # по порядковому номеру (всегда работает)
# или конкретное значение:
df.iloc[4]['temp']      # сначала строка, потом колонка
# или через .at для скалярного значения (быстрее):
df.at[4, 'temp']        # по метке индекса
df.iat[4, 5]            # по номеру строки и колонки
```

## 16. Срезы через `.loc` и `.iloc`

**Задача:** Получи строки с `datetime` от `'2011-01-01 01:00:00'` до `'2011-01-01 03:00:00'` и каждую вторую строку с позиции 10 по 20.  
**Подсказка:** `.loc[start:end]` и `.iloc[start:end:step]`.

**Если забыл `.loc`-срезы:**
```python
# Через булеву фильтрацию:
mask = (df['datetime'] >= '2011-01-01 01:00:00') & (df['datetime'] <= '2011-01-01 03:00:00')
df[mask]

# Для iloc-среза — обычный Python-срез:
df_dt[10:20:2]
```

## 17. Изменение значения через `.loc`

**Задача:** Измени значение `weather` для строки с индексом `'2011-01-01 04:00:00'` на 234.  
**Подсказка:** `.loc[индекс, колонка] = значение`.

**Если забыл `.loc` для записи:**
```python
df_dt.at['2011-01-01 04:00:00', 'weather'] = 234     # .at — быстрее для скалярных значений
# ВАЖНО: df_dt.iloc[4]['weather'] = -5 — НЕ гарантирует запись (SettingWithCopyWarning)!
# Всегда используй .loc или .at для записи.
```

## 18. Сброс индекса — `.reset_index()`

**Задача:** Верни индекс обратно в колонку (сбрось на 0, 1, 2…).  
**Подсказка:** `.reset_index()`.

**Если забыл `.reset_index()`:**
```python
# Вручную:
df_dt['datetime'] = df_dt.index       # скопировать индекс в колонку

```

## 19. Группировка — `.groupby().agg()`

**Задача:** Посчитай среднюю температуру в разбивке по сезонам.  
**Подсказка:** `.groupby('col').agg({'col': 'func'})`.

**Если забыл `.agg()`:**
```python
df.groupby('season')['temp'].mean()           # вызвать mean() напрямую
# или через цикл (когда совсем забыл pandas):
for season in df['season'].unique():
    subset = df[df['season'] == season]
    print(f'Сезон {season}: средняя t = {subset["temp"].mean():.2f}')
```

---
## 20. Группировка по нескольким колонкам

**Задача:** Посчитай среднюю влажность в разбивке по сезонам **и** рабочим дням.  
**Подсказка:** `.groupby([col1, col2])['col'].mean()`.

**Если забыл `.groupby()` с несколькими колонками:**
```python
# Через pivot_table — часто удобнее для двух группировок:
df.pivot_table(values='humidity', index='season', columns='workingday', aggfunc='mean')
```

## 23. Группировка по части даты

**Задача:** Посчитай среднюю температуру и минимальную влажность **по месяцам**.

df_1['month'] = df_1['datetime'].dt.month
df_1.groupby('month').agg({'temp': 'mean', 'humidity': 'min'})
# Потом можно удалить: df_1.drop(columns='month', inplace=True)
```

df_1.set_index('datetime').resample('D').count()
```

## 27. Точная настройка через matplotlib

**Задача:** Построй линию + точки на одном графике (понедельные поездки).

## Шпаргалка: все методы одним списком

| #  | Задача | Метод | Альтернатива |
|----|--------|-------|--------------|
| 1  | Чтение CSV | `pd.read_csv()` | `csv.DictReader` + `pd.DataFrame` |
| 2  | Первые строки | `.head(n)` | `df[:n]` |
| 3  | Статистика | `.describe()` | `.mean()`, `.std()`, `.min()`, `.max()` |
| 4  | Колонки | `.columns` | `list(df)` |
| 5  | Типы | `.dtypes` | `.info()` |
| 6  | Фильтрация | `df[df['col'] == val]` | `df.query('col == val')` |
| 7  | И | `&` | `query('a and b')` |
| 8  | ИЛИ | `\|` | `query('a or b')` |
| 9  | НЕ | `~` | `!=` или `query('not ...')` |
| 10 | Пропуски | `.isna()` | `.isnull()` |
| 11 | В списке? | `.isin([...])` | `query('col in [...]')` |
| 12 | По индексу | `.loc[]` / `.iloc[]` | `.at[]` / `.iat[]` |
| 13 | Сменить индекс | `.set_index()` | `df.index = ...` |
| 14 | Сбросить индекс | `.reset_index()` | `df.index = range(len(df))` |
| 15 | Группировка | `.groupby().agg()` | `.groupby()['col'].mean()` |
| 16 | В дату | `pd.to_datetime()` | `parse_dates=` в `read_csv` |
| 17 | Часть даты | `.dt.month` и т.д. | `.apply(lambda x: x.month)` |
| 18 | График | `.plot()` | `plt.plot()` |
| 19 | Гистограмма | `.hist()` | `plt.hist()` |
| 20 | В CSV | `.to_csv()` | `csv.DictWriter` |
| 21 | В Excel | `.to_excel()` | `.to_csv()` → открыть в Excel |
